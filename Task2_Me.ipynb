```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hp= pd.read_csv("/content/gdrive/MyDrive/Task2hp/HPSpells.csv")
```

```python
hp.head()
```

```python
hp.drop(hp[hp.Type == 'Hex'].index,inplace =True)
```

```python
hp.drop(hp[hp.Light == 'NaN'].index,inplace =True)
```

```python
freq= {}
for Type in hp.Type.unique():
  c=0
  for item in hp.Type:
    if item ==Type:
      c+=1
    freq[Type]=c
plt.figure(figsize=(10,7))
plt.bar(range(len(hp.Type.unique())),list(freq.values()),color = 'grey')
plt.xticks(range(len(freq)),hp.Type.unique())
plt.show() 
```
